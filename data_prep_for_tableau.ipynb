```python
import yfinance as yf
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

# Download data
tickers = ['TLT', 'IEF', 'SHY']
data = yf.download(tickers, period="5y", auto_adjust=True)['Close']
returns = data.pct_change().dropna()

# Portfolio returns
portfolio_returns = returns.mean(axis=1)

# Rolling metrics
rolling_vol = portfolio_returns.rolling(30).std()
var_95_rolling = portfolio_returns.rolling(30).quantile(0.05)

# Risk flag
risk_flag = np.where(portfolio_returns < var_95_rolling, 'HIGH RISK', 'Normal')

# Final dataset
df = pd.DataFrame({
    'Date': portfolio_returns.index,
    'Portfolio_Return': portfolio_returns,
    'Rolling_Vol_30d': rolling_vol,
    'VaR_95_30d': var_95_rolling,
    'Risk_Flag': risk_flag,
    '+50bps_Stress_Return': portfolio_returns * 1.25,
    '+100bps_Stress_Return': portfolio_returns * 1.5
})

# Save
df.to_csv('market_risk_dashboard_data.csv', index=False)

print("File created successfully!")
df.head()
```
